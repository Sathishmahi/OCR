```python
import keras_ocr
```

```python
img_arr=keras_ocr.tools.read("1.jpg")
```

```python
img_arr.shape
```

```python
pipeline=keras_ocr.pipeline.Pipeline()
```

```python
result=pipeline.recognize([img_arr])
```

```python
result[0][0]
```

```python
import cv2
imarr= cv2.imread("1.jpg")
from google.colab.patches import cv2_imshow
for res in result[0]:
  p1,p2 = (int(res[1][0][0]),int(res[1][0][1])),(int(res[1][2][0]),int(res[1][2][1]))
  text = res[0]
  cv2.rectangle(imarr,p1,p2,(255,0,0),1,)
  cv2.putText(imarr,text,p1,cv2.FONT_HERSHEY_COMPLEX,0.5,(0,0,0),1)
```

```python
cv2_imshow(imarr)
```
