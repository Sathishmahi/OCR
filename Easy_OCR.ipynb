```python
import easyocr
reader=easyocr.Reader(["en"],gpu=False)
```

```python
result=reader.readtext("1.jpg")
```

```python
result[0][1]
```

```python
p1=tuple(result[0][0][0])
p2=tuple(result[0][0][2])
```

```python
import cv2
from google.colab.patches import cv2_imshow
imarr=cv2.imread("1.jpg")

for res in result:
  p1=tuple(res[0][0])
  p2=tuple(res[0][2])
  text=res[1]
  cv2.rectangle(imarr,p1,p2,(255,0,0),2)
  cv2.putText(imarr,text,(int(p1[0]),int(p1[1]-10)),cv2.FONT_HERSHEY_DUPLEX,0.5,(0,0,0),1)
```

```python
cv2_imshow(imarr)
```
